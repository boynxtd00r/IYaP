### Задание 1


```python
from functools import reduce
from typing import List

def word_maker(letters: List[str]) -> str:
    word = reduce(lambda a, b: a + b.upper(), letters, '')
    return word
```

```python
def word_maker_2(letters: List[str]) -> str:
    word = ''
    for l in letters:
        word += l.upper()
    return word
```

```python
print(word_maker(('a', 'b', 'c')))
print(word_maker_2(('x', 'y', 'z')))
```

### Задание 2


```python
from typing import List

def filter_sized(words: List[str], n: int) -> List[str]:
    f = sorted([w for w in words if len(w) == n])
    return f
```

```python
print(filter_sized(['дом', 'улица', 'гараж', 'работа', 'он', 'она'], 3))
```

### Задание 3


```python
import datetime
from typing import Tuple

def dmy(date: datetime.datetime) -> Tuple[int, int, int]:
    if not isinstance(date, datetime.datetime):
        raise TypeError
    
    d = (lambda x: x.day)(date)
    m = (lambda x: x.month)(date)
    y = (lambda x: x.year)(date)
    
    return d, m, y
```

```python
dmy(datetime.datetime.now())
```

### Задание 4


```python
def intersection(s1, s2):
    l1 = s1.split()
    l2 = s2.split()
    
    res = list(filter(lambda x: x in l2, l1))
    
    return ' '.join(res)
```

```python
intersection('1 2 3 4 5', '3 4 5 6 7 8')
```

```python
def check_password(password: str, min_length: int):
    is_strong = (
        len(password) >= min_length and
        any(c.isupper() for c in password) and
        any(c.islower() for c in password) and
        any(c.isdigit() for c in password) and
        any(c in "!@#$%^&*()-+" for c in password)
    )
    return is_strong
```

```python
print(check_password('1aB!2', 5))
print(check_password('12345', 5))
```

```python
def check_password_error(password: str, min_length: int):
    problems = []
    
    if not any(c.isupper() for c in password):
        problems.append('заглавная')
        
    if not any(c in "!@#$%^&*()-+" for c in password):
        problems.append('символ')
        
    if not any(c.islower() for c in password):
        problems.append('прописная')
        datetime.datetime.now()
    if not any(c.isdigit() for c in password):
        problems.append('цифра')
        
    if len(password) < min_length:
        problems.append(f'длина {len(password)} символов из {min_length}')
        
    if problems:
        return False, problems
    
    return True
```
