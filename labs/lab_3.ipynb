# Работа со списками

```python
matrix = [
    [1, 2],
    [5, 6],
    [9, 10],
]
t_matrix = []
elem_t_matrix = []
for i in range(len(matrix[0])):
    for j in range(len(matrix)):
        elem_t_matrix.append(matrix[j][i])
    t_matrix.append(elem_t_matrix)
    elem_t_matrix = []
print(t_matrix)
```

# Работа со множествами

```python
s1 = {1, 2, 3, 4, 5, 6, 7}
s2 = {5, 6, 7, 8, 9}
s3 = s1 - s2
s4 = s2 - s1
s5 = s1 & s2
print(s3, s4, s5)
```

# Работа со словарями

```python
s1 = {1: 1, 2: 1, 3: 1, 4: 1, 5: 1, 6: 1, 7: 1}
s2 = {5: 1, 6: 1, 7: 1, 8: 1, 9: 1}
s3 = {}
s4 = {}
s5 = {}

for i in s1:
    if i in s1 and i in s2:
        s3[i] = 1

for i in s1:
    if i not in s2:
        s4[i] = 1

for i in s2:
    if i not in s1:
        s5[i] = 1

print(s3.keys())
print(s4.keys())
print(s5.keys())
```

# Задача "Рекомендации"

```python
from typing import List

recom_ids = [1, 1, 2, 3, 3]
seen_ids = [1, 4, 5]

def get_filtered(recom_ids: List[int], seen_ids: List[int]) -> List[int]:
    filtered = []
    for i in recom_ids:
        if i not in seen_ids:
            filtered.append(i)
    return filtered

print(get_filtered(recom_ids, seen_ids))
```

# Задача "Поиск максимальных элементов в массиве"

```python
from typing import List

nums = [1,2,3,2,5,7,1]
k = 2

def get_top(nums: List[int], k: int):
    top = []
    for i in range(k):
        maks = max(nums)
        top.append(maks)
        nums.pop(nums.index(maks))
        maks = []
    return top

print(get_top(nums, k))
```

# Задача "Поиск часто встречающихся элементов в массиве"

```python
nums = [1,2,3,3,3,3,3,2,5,7,1]
k = 2
elems = set(nums)

from typing import List

def get_top(nums: List[int], k: int):
    top = []
    for i in range(k):
        counter = 0
        top_elem = 0
        for elem in elems:
            if counter < nums.count(elem):
                counter = nums.count(elem)
                top_elem = elem
        top.append(top_elem)
        elems.discard(top_elem)
    return top

print(get_top(nums, k))
```
