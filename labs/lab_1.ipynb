```python
def main():
    '''Выводится фраза "Hello, world!" через f-строку и проверяется тип x'''
    x = '1'

    name = "World"
    print(f"Hello, {name}!")

    match x:
        case int(x): print("Это число")
        case str(x): print("Это строка")


if __name__ == '__main__':
    main()    
```
