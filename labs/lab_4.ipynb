## Работа с аргументами

```python
from functools import wraps

def unexcept(fun):
    """
    Этот декоратор не позволяет функции возражать исключения
    """
    @wraps(fun)
    def wrapped(*args, **kwargs):
        e = None
        try:
            res = fun(*args, **kwargs)
        except Exception as ex:
            res, e = 0, ex
        return res, e

    return wrapped

@unexcept
def function(operation, start, *args):
    result = start
    
    for num in args:
        if operation == '+':
            result += num
        elif operation == '-':
            result -= num
        elif operation == '*':
            result *= num
        elif operation == '/':
            result /= num
    
    return result

print(function('-', 6, 2, 1))
print(function('-', 6, 10))
print(function('+', 5, 1, 2, 3))
print(function('*', 2, 3, 4))
print(function('/', 10, 5))
print(function('/', 10, 0))
```

## Декораторы

#### 1 Задача

```python
from functools import wraps

def unexcept(default):
    """
    Декоратор который получает на вход обязательный аргумент.
    В случае если в декорируемой функции возникнет исключение, функция вернет аргумент, получаемый декоратором.
    """
    def wrapper(func):
        @wraps(func)
        def wrapped(*args, **kwargs):
            try:
                return func(*args, **kwargs)
            except Exception as e:
                print(f"Поймано исключение: {e}")
                return default
        return wrapped
    return wrapper

@unexcept(default=-1)
def div(a, b):
    return a / b

@unexcept(default=1000)
def plus(a, b):
    return a + b

print(div(10, 2))
print(div(10, 0))
print(plus(5, 3))
print(plus(5, '3'))
```

#### 2  Задача

```python
from time import time
from functools import wraps

def n_dec(n):
    """
    Декоратор, выполняющий функцию n раз и выводящий общее время выполнения.
    """
    def wrapper(func):
        @wraps(func)
        def wrapped(*args, **kwargs):
            start_time = time()
            
            for i in range(n):
                result = func(*args, **kwargs)
            
            end_time = time()
            print(f"Функция выполнена {n} раз за {end_time - start_time} секунд")
            return result
        return wrapped
    return wrapper

@n_dec(n=1000000)
def plus(a, b):
    return a + b

plus(5, 3)
```

## Дебаг

```python
def free_ports(min_port, max_port, busy: list):
    result = []

    busy.sort()

    range_start = min_port

    for port in busy:
        if port > max_port:
            continue
        if port - range_start > 0:
            result.append([range_start, port - 1])
        range_start = port + 1
        if range_start < min_port:
            range_start = min_port

    if range_start < max_port:
        result.append([range_start, max_port])

    return result

print(free_ports(30000, 32000, [30100, 30200]))
print(free_ports(30100, 30200, [30100, 30200]))
print(free_ports(30000, 32000, [32100]))
```

### 2  Задача

```python
import string

def is_pangram(text, alphabet=string.ascii_lowercase):
    text = text.lower()
    letters = set(text)
    
    return set(alphabet).issubset(letters)

test_strings = [
    "The quick brown fox jumps over the lazy dog",
    "Hello world"
]

for s in test_strings:
    print(f"'{s}' -> {is_pangram(s)}")
```
