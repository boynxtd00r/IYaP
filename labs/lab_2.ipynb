## Задание 2 (Реверс строки)


```python
text = input("Введите три слова через пробел: ")

#Способ 1
words = text.split()
reversed_text = '_'.join(reversed(words))
print(f"Способ 1: {reversed_text}")

#Способ 2
words = text.split()
reversed_text = f"{words[2]}_{words[1]}_{words[0]}"
print(f"Способ 2: {reversed_text}")
```

# Арифметические операции

## Задание 1 (Чаевые в ресторане)

```python
bill = int(input("Введите сумму счёта: "))
tip_percent = int(input("Введите процент чаевых: "))

tip_amount = bill * tip_percent / 100
total = bill + tip_amount

print(f"Счёт: {bill} руб. Чаевые ({tip_percent}%): {tip_amount} руб. Итого: {total} руб.")
```

## Задание 2 (Площадь кольца)

```python
import math

R = int(input("Введите внешний радиус: "))
r = int(input("Введите внутренний радиус: "))

area = math.pi * (R**2 - r**2)

print(f"Площадь кольца с радиусами {R} и {r} = {area}")
```

# Работа со строками

## Задание 1 (Генератор никнеймов)

```python
full_name = input("Введите имя и фамилию через пробел: ")

name_parts = full_name.split()
first_name = name_parts[0]
last_name = name_parts[1]
nickname = last_name[:3].upper() + "_" + first_name[:2].lower() + str(len(full_name))
    
print(f"Никнейм: '{nickname}'")
```

## Задание 2 (Поиск хэштега)

```python
text = input("Введите текст с хэштегом: ")

words = text.split()

for word in words:
    if word[0] == "#":
        hashtag = word[1:].upper()
        print(hashtag)
        break
```

## Задание 1

```python
counter = 0

#Пример 1
print("Пример 1: ")

s = 5
while str(True) and s >= 0:
    s -= 1
    counter += 1
    
print(f"Всего {counter} итераций.")
counter = 0

#Пример 2
print("Пример 2: ")

s = 5
while str(False) and s >= 0:
    s -= 1
    counter += 1

print(f"Всего {counter} итераций.")
counter = 0

#Пример 3
print("Пример 3: ")

s = 5
while str(None) and s >= 0:
    s -= 1
    counter += 1

print(f"Всего {counter} итераций.")
```

## Задание 2 ("Угадай число")

```python
secret = 42
attempts = 5

print("Угадай число!")

for attempt in range(1, attempts + 1):
    guess = int(input(f"Попытка {attempt}: "))
    
    if guess == secret:
        print("Поздравляю!")
        break
    elif guess < secret:
        print("Больше")
    else:
        print("Меньше")
else:
    print(f"Попытки кончились. Было загадано: {secret}")
```

##  Задание 3 (Сумма цифр числа)

```python
num = int(input("Введите целое неотрицательное число: "))

sum_digits = 0
n = num

while n > 0:
    sum_digits += n % 10
    n = n // 10

print(f"Сумма цифр числа {num}: {sum_digits}")
```
